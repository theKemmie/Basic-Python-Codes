```python
# A Python program for displaying all prime numbers within a certain range

#Get numbers
first = int(input("Please enter the first number in the interval:  "))
last = int(input("Please enter the last number in the interval:  "))

print("Prime numbers between", first, "and", last, "are:")

for num in range(first, last + 1):
   # all prime numbers are greater than 1
   if num > 1:
       for i in range(2, num):
           if (num % i) == 0:
               break
       else:
           print(num ,"\n")
else:
    print("No prime numbers in this interval")
```
